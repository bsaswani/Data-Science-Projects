# Pandas Project 4 - Laptop Price

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
df=pd.read_csv('laptop_prices.csv')
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# 1. Which laptop Company Sells high Priced laptops?
a = df.groupby('Company')['Price_euros'].max().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=a, x='Company', y='Price_euros', palette="viridis")
plt.xticks(rotation=45)
plt.title('Highest Priced Laptop by Company')
plt.xlabel('Company')
plt.ylabel('Highest Price (Euros)')
```

```python
# 2. Which laptop company have giving lowest Touch Screen feature in their products?
t=df.query("Touchscreen=='Yes'")
plt.figure(figsize=(15, 6))
sns.countplot(data=t, x="Company")
plt.xticks(rotation=45)
plt.title('Count of Laptops with Touchscreen Feature by Company')
plt.xlabel('Company')
plt.ylabel('Count')
```

```python
# 3. Which company has giving a maximum type of '2 in 1 Convertible' laptops?
t=df.query("TypeName=='2 in 1 Convertible'")
c=t['Company'].value_counts()
plt.figure(figsize=(15,6))
p=plt.bar(x=c.index,height=c.values,color='g',edgecolor='y')
plt.bar_label(p, padding=3, fontsize=10)
plt.xticks(rotation=45)
plt.title('Count of 2 in 1 Convertible Laptops by Company')
plt.xlabel('Company')
plt.ylabel('Count')
```

```python
# 4. How many price is charged for Screen 4k UltraHD for each company? 
u=df.query("Screen=='4K Ultra HD'")
a=u.groupby('Company')['Price_euros'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=a, x='Company', y='Price_euros', palette='coolwarm')
plt.title('Total Price Charged for 4K Ultra HD Screens by Company')
plt.xlabel('Company')
plt.ylabel('Total Price (Euros)')
```

```python
# 5. Which Company sells their laptop without any operating system?
df['OS'].unique()
u=df.query("OS=='No OS'")
a=u.groupby(['Company','Product'])['Price_euros'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=a, x='Company',y='Price_euros',hue='Product',palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Total Price Charged for laptop without any OS by Company')
plt.xlabel('Company')
plt.ylabel('Total Price (Euros)')
plt.legend(title='Product', bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
# 6. Which Company has largest display Screen?
d=df.groupby('Company')['Inches'].max().reset_index()
plt.figure(figsize=(15,6))
p=plt.bar(d['Company'],d['Inches'],color='Salmon',edgecolor='r')
plt.bar_label(p, padding=3, fontsize=10)
plt.xticks(rotation=45)
plt.title('Maximum Display Screen Size by Company')
plt.xlabel('Company')
plt.ylabel('Screen Size (Inches)')
```

```python
# 7. How many product are selled by company Toshiba?
c=df.query("Company=='Toshiba'")
#c['Product'].value_counts()
plt.figure(figsize=(10, 6))
sns.countplot(data=c, x="Product",palette='Spectral')
plt.xticks(rotation=90)
plt.title('Count of Products by Company Toshiba')
plt.xlabel('Products')
plt.ylabel('Count')
```

```python
# 8. Name the company and their total cost  has selling product with standard screen?
s=df.query("Screen=='Standard'")
a=s.groupby('Company')['Price_euros'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=a, x='Company', y='Price_euros', palette='hsv')
plt.title('Total Price Charged for Standard Screens by Company')
plt.xlabel('Company')
plt.ylabel('Total Price (Euros)')
```

```python
# 9. Name the product and its price which have GPU model GeForce GTX 965M
s=df.query("GPU_model=='GeForce GTX 965M'")
a=s.groupby('Product')['Price_euros'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=a, x='Product', y='Price_euros', palette='mako')
plt.title('Price Charged for GPU model GeForce GTX 965M by Product')
plt.xlabel('Product')
plt.ylabel('Total Price (Euros)')
```

```python
# 10. Which company highest variants of Gaming laptops?
a=df.query("TypeName=='Gaming'")
h=a["Company"].value_counts()
c=plt.bar(x=h.index,height=h.values,color='r',edgecolor='y')
plt.bar_label(c, padding=3, fontsize=10)
plt.title("Count of Highest Gaming Laptop Variants by Company")
plt.xlabel("Company")
plt.ylabel("Number of Gaming Laptops in each company")
```

```python
# 11. Which company has light weight laptops available?
w=df.query("Weight<1.0")
plt.figure(figsize=(10, 6))
sns.barplot(data=w, x='Company',y='Weight',palette='Oranges')
plt.title("Count of Company that has light weight laptops")
plt.xlabel("Company")
plt.ylabel("Weights (kg)")
```

```python
# 12. Which laptop Products are available for screen inch below 15 inches and company name Apple?
a=df.query("Inches<15 and Company=='Apple'")
i=a["Product"].value_counts()
c=plt.bar(x=i.index,height=i.values,color='Teal',edgecolor='y')
plt.bar_label(c, padding=3, fontsize=10)
plt.title("Laptop Product available for screen inch below 15 in Apple")
plt.xlabel("Product")
plt.ylabel("Number of Laptops for each company")
```

```python
# 13. Which company has the laptops having IPSpanel included?
a=df.query("IPSpanel=='Yes'")
ips=a["Company"].value_counts()
plt.figure(figsize=(15, 6))
c=plt.bar(x=ips.index,height=ips.values,color='Lime',edgecolor='g')
plt.bar_label(c, padding=3, fontsize=10)
plt.title("Laptop Product With IPS Panel")
plt.xlabel("Product")
plt.ylabel("Number of Laptops for each company")
```

```python
# 14. What will be laptop that having lowest price for the AMD?
a=df.query("CPU_company=='AMD'")
ips=a.groupby(['Company','Product'])['Price_euros'].min().reset_index()
plt.figure(figsize=(35, 8))
sns.barplot(data=ips, x='Product', y='Price_euros', hue='Company', palette='Blues')
plt.title('Laptops with Lowest Prices for AMD CPUs', fontsize=16)
plt.xlabel('Product', fontsize=14)
plt.ylabel('Price (Euros)', fontsize=14)
plt.xticks(rotation=90)
```

```python
# 15. Which laptop company provides the Hybrid primary storage type? 
ps=df.query("PrimaryStorageType=='Hybrid'")
cps=ps['Company'].value_counts()
plt.figure(figsize=(10, 8))
c=plt.bar(x=cps.index,height=cps.values,color='Lavender',edgecolor='g')
plt.bar_label(c, padding=3, fontsize=10)
plt.title('Laptops with Hybrid Primary Storage', fontsize=16)
plt.xlabel('Company', fontsize=14)
plt.ylabel('Number of Laptops in each Company', fontsize=14)
```

```python
# 16. What will the highest price for the Core i7 6500U?
a=df.query("CPU_model=='Core i7 6500U'")
i7=a.groupby(['Product','Company'])['Price_euros'].max().reset_index()
plt.figure(figsize=(35,12))
sns.barplot(data=i7, x='Product', y='Price_euros',hue='Company',palette='flare')
plt.title('Laptops with Highest Prices for Core i7 6500U', fontsize=16)
plt.xlabel('Product', fontsize=18)
plt.ylabel('Price (Euros)', fontsize=18)
plt.xticks(rotation=90)
```

```python
# 17. Which company provides the laptop screen as Quad HD+?
qs=df.query("Screen=='Quad HD+'")
qshd=qs['Company'].value_counts()
plt.figure(figsize=(10, 8))
c=plt.bar(x=qshd.index,height=qshd.values,color='Orchid',edgecolor='r')
plt.bar_label(c, padding=3, fontsize=10)
plt.title('Laptops with Quad HD+ Screen', fontsize=16)
plt.xlabel('Company', fontsize=14)
plt.ylabel('Number of Laptops in each Company', fontsize=14)
```

```python
# 18. What is the Company/Price that provides the Retina Display? 
df["RetinaDisplay"].unique()
a=df.query("RetinaDisplay=='Yes'")
rd=a.groupby(['Company','Product'])['Price_euros'].sum().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=rd, x='Company', y='Price_euros',hue='Product',palette='inferno')
plt.title('Laptops with Retina Display', fontsize=16)
plt.xlabel('Company', fontsize=18)
plt.ylabel('Price (Euros)', fontsize=18)
```

```python
# 19. What is Minimum price for laptop that having linux operating system? 
a=df.query("OS=='Linux'")
los=a.groupby(['Company','Product'])['Price_euros'].min().reset_index()
plt.figure(figsize=(30,6))
sns.barplot(data=los, x='Product', y='Price_euros',hue='Company',palette='rocket')
plt.title('Laptops Price with Linux Operating System', fontsize=16)
plt.xlabel('Product', fontsize=18)
plt.ylabel('Price (Euros)', fontsize=18)
plt.xticks(rotation=90)
```
