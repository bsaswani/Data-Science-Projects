# Pandas Project 3 - IPL Complete Dataset (2008-2024)

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
df=pd.read_csv('matches.csv')
df
```

```python
df.info()
```

```python
df.columns
```

```python
# checking null values
df.isnull().sum()
```

```python
# filling and replace null values
df['method'].fillna('Normal', inplace=True)
df['player_of_match'].fillna('No Player',inplace=True)
df['winner'].fillna('No Winner',inplace=True)
df['result_margin'].fillna(0.0,inplace=True)
df.drop(df[df['id'] == 501265].index, inplace=True)
df.drop(df[df['id'] == 829763].index, inplace=True)
df.drop(df[df['id'] == 1359519].index, inplace=True)

df['city']=df['city'].fillna(df['venue'].apply(lambda x: 'Dubai' if x == 'International Cricket Stadium' else ('Sharjah' if x == 'Sharjah Cricket Stadium' else 'default')))
```

```python
df.isnull().sum()
```

```python
df.duplicated()
```

```python
# 1. Which Team has most of matches are won in the Final?
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
a=df.query("match_type=='Final'")
b=a['winner']
plt.figure(figsize=(10, 6))
sns.countplot(x=b,palette='cividis')
plt.title('Teams with Most Wins in Finals')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=90)
```

```python
# 2. Which Team has loss in most of semi final round?
a=df.query("match_type=='Semi Final'")
a['loser']=a.apply(lambda x: x['team1'] if x['team2'] == x['winner'] else x['team2'], axis=1)
loss_count=a['loser'].value_counts()
plt.figure(figsize=(10, 6))
c=plt.bar(x=loss_count.index,height=loss_count.values,color='g',edgecolor='y')
plt.bar_label(c, padding=3, fontsize=10)
plt.title('Teams with Most Losses in Semi-Final Matches')
plt.xlabel('Teams')
plt.ylabel('Number of Losses')
plt.xticks(rotation=45)
```

```python
# 3. How many matches are played on the Venue in M Chinnaswamy Stadium ?
s=df.query("venue=='M Chinnaswamy Stadium'")
s['venue'].count()
```

```python
# 4. Which is the most common toss winner teams in ipl?
toss_win=df['toss_winner'].value_counts()
plt.figure(figsize=(15, 6))
d=plt.bar(toss_win.index,toss_win.values,color='purple',edgecolor='r')
plt.bar_label(d,padding=3, fontsize=10)
plt.title("Most Common Toss Winner Teams")
plt.xlabel("Teams")
plt.ylabel("Number of Toss Wins")
plt.xticks(rotation=90)
```

```python
# 5. Plot the teams that has lost finals in IPL?
a=df.query("match_type=='Final'")
a['loser']=a.apply(lambda x: x['team1'] if x['team2'] == x['winner'] else x['team2'], axis=1)
b=a['loser'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=b.index,y=b.values,palette='magma')
plt.title('Teams with Most Lost in Finals')
plt.xlabel('Team')
plt.ylabel('Number of Lost')
plt.xticks(rotation=90)
```

```python
# 6. Which teams are won through a super over?
s=df.query("super_over=='Y'")
super_over=s['winner'].value_counts()
plt.figure(figsize=(10, 6))
d=plt.bar(super_over.index,super_over.values,color='Coral',edgecolor='r')
plt.bar_label(d,padding=3, fontsize=10)
plt.title("Most Common Team Won Through Super Over")
plt.xlabel("Teams")
plt.ylabel("Number of Super Over Wins")
plt.xticks(rotation=90)
```

```python
# 7. Plot the minimum match overs in the ipl? 
a=df.query("target_overs<15")
min_over=a['target_overs'].value_counts()
plt.figure(figsize=(15, 6))
d=plt.bar(min_over.index,min_over.values,color='Lime',edgecolor='g')
plt.bar_label(d,padding=3, fontsize=10)
plt.title("Minimum Match Over in IPL")
plt.xlabel("Overs")
plt.ylabel("Number of Minimum Overs")
plt.xticks(rotation=90)
```

```python
# 8. Plot top 15 players that wons the most player of the match?
player=df['player_of_match'].value_counts().head(15)
plt.figure(figsize=(15, 6))
d=plt.bar(player.index,player.values,color='Salmon',edgecolor='r')
plt.bar_label(d,padding=3, fontsize=10)
plt.title("Top Most Man of the Match In IPL")
plt.xlabel("Players")
plt.ylabel("Number of Man of the Match")
plt.xticks(rotation=90)
```

```python
# 9. Which is the minimum toss decision make by teams?
toss_counts = df['toss_winner'].value_counts().sort_values(ascending=True)
plt.figure(figsize=(15, 6))
d=plt.bar(toss_counts.index,toss_counts.values,color='Turquoise',edgecolor='g')
plt.bar_label(d,padding=3, fontsize=10)
plt.title("Minimum Toss Decision Taken By Teams")
plt.xlabel("Teams")
plt.ylabel("Number of Toss Decision")
plt.xticks(rotation=90)
```

```python
# 10. Which is the lowest target runs in ipl?
a=df.query("target_runs<100")
min_run=a['target_runs'].value_counts().sort_values(ascending=True)
plt.figure(figsize=(15, 6))
d=plt.bar(min_run.index,min_run.values,color='Gold',edgecolor='r')
plt.bar_label(d,padding=3, fontsize=10)
plt.title("Minimum Match Runs in IPL")
plt.xlabel("Runs")
plt.ylabel("Number of Minimum Runs")
plt.xticks(rotation=90)
```

```python
# 11. How many matches are ended as Tie?
t=df.query("result=='tie'")
t['result'].value_counts()
```

```python
# 12. Which is team who most lost in the ipl in final upto season 2013?
a=df.query("match_type=='Final' and season <'2014'")
a['loser']=a.apply(lambda x: x['team1'] if x['team2'] == x['winner'] else x['team2'], axis=1)
b=a['loser'].value_counts()
plt.figure(figsize=(8, 6))
palette=sns.color_palette("pastel")
sns.barplot(x=b.index,y=b.values,palette=palette)
plt.title('Teams with Most Lost in Finals')
plt.xlabel('Team')
plt.ylabel('Number of Lost')
```

```python
# 13. Plot the top 7 stadiums for the matches played in ipl?
venue_counts = df['venue'].value_counts().head(7)
plt.figure(figsize=(15, 6))
d=plt.bar(venue_counts.index,venue_counts.values,color='MistyRose',edgecolor='r')
plt.bar_label(d,padding=3, fontsize=10)
plt.title("Top 7 most match played stadium in IPL")
plt.xlabel("Stadium")
plt.ylabel("Number of Match Played")
plt.xticks(rotation=45)
```

```python
# 14. Count the matches that are played Duckworth-Lewis (D/L) method?
d=df.query("method=='D/L'")
d['method'].value_counts()
```

```python
# 15. Display the teams who won the match using Duckworth-Lewis method in ipl?
d=df.query("method=='D/L'")
dl=d['winner'].value_counts()
plt.figure(figsize=(15, 6))
a=plt.bar(dl.index,dl.values,color='PeachPuff',edgecolor='r')
plt.bar_label(a,padding=3, fontsize=10)
plt.title("Teams won using D/L method in IPL")
plt.xlabel("Teams")
plt.ylabel("Number of Match Won")
plt.xticks(rotation=45)
```

```python
# 16. Plot the matches which is won by wicket and runs?
r=df.query("result=='runs'")
runs=r['result'].value_counts()
w=df.query("result=='wickets'")
wickets=w['result'].value_counts()
a=plt.bar(runs.index,runs.values,color='Goldenrod',edgecolor='r')
b=plt.bar(wickets.index,wickets.values,color='Gold',edgecolor='r')
plt.bar_label(a,padding=3, fontsize=10)
plt.bar_label(b,padding=3, fontsize=10)
plt.title("Teams won by Wicket/ Runs in IPL")
plt.xlabel("Result")
plt.ylabel("Number of Match Won")
```

```python
# 17. Plot the highest of runs margin wins matches played in ipl?
a=df.query("result=='runs'")
margin=a['result_margin'].value_counts().sort_values(ascending=True).head(10)
plt.figure(figsize=(12, 6))
plt.bar(margin.index, margin.values, color='skyblue', edgecolor='black')
plt.title("Highest Run Margin Wins in IPL", fontsize=16)
plt.xlabel("Run Margin", fontsize=14)
plt.ylabel("Number of Matches Won", fontsize=14)
plt.xticks(rotation=45) 
```

```python
# 18. Which team is least choosed toss decision as to bat or field?
b=df.query("toss_decision=='bat'")
batting=b['toss_decision'].value_counts()
f=df.query("toss_decision=='field'")
fielding=f['toss_decision'].value_counts()
a=plt.bar(batting.index,batting.values,color='MistyRose',edgecolor='r')
b=plt.bar(fielding.index,fielding.values,color='LightCoral',edgecolor='r')
plt.bar_label(a,padding=3, fontsize=10)
plt.bar_label(b,padding=3, fontsize=10)
plt.title("Teams who choose for batting / fielding IPL")
plt.xlabel("Toss Decision")
plt.ylabel("Number of batting / fielding")
```

```python
# 19. Which city played a least game while comparing all season?
city_counts = df['city'].value_counts().sort_values().head(10)
plt.figure(figsize=(15, 6))
d=plt.bar(city_counts.index,city_counts.values,color='y',edgecolor='r')
plt.bar_label(d,padding=3, fontsize=10)
plt.title("Least match played city in IPL")
plt.xlabel("city")
plt.ylabel("Number of Match Played")
plt.xticks(rotation=45)
```
