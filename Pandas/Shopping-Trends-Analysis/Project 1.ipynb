```python
# 1. What is the most common season for purchases ?
import matplotlib.pyplot as plt
import numpy as np
x=df['Season'].value_counts()
plt.xlabel('Season')
plt.ylabel('Number of Customer')
plt.title('Most common season for purchase')
a=plt.bar(x.index,x.values,color='c',edgecolor='k')
plt.bar_label(a,padding=2,fontsize=10)
```

```python
# 2. What is the total purchase amount for each gender ?
import matplotlib.pyplot as plt
import numpy as np
x=df.groupby('Gender')['Purchase Amount (USD)'].sum()
plt.xlabel('Gender')
plt.ylabel('Number of Customer')
plt.title('Total purchase amount')
a=plt.bar(x.index,x.values,color='g',edgecolor='k')
plt.bar_label(a,padding=2,fontsize=10)
```

```python
#3. What is the average age of customers who made purchases in the Summer season ?
cust=df.query("Season=='Summer'")
cust['Age'].mean()
```

```python
# 4. How many customers used a promo code for their purchase ?
cust_promo=df.query("`Promo Code Used`=='Yes'")
a = cust_promo['Promo Code Used'].count()
a
```

```python
# 5. What is the maximum and minimum review rating in the dataset ?
a=df['Review Rating'].min()
b=df['Review Rating'].max()
print("Minimum Review Rating: ",a)
print("Maximum Review Rating: ",b)
```

```python
# 6. What is the most common payment method for customers who shop in the Winter season ?
import matplotlib.pyplot as plt
import numpy as np
cust1=df.query("Season=='Winter'")
x=cust1['Payment Method'].value_counts()
plt.xlabel('Payment Method')
plt.ylabel('Winter Season')
plt.title('Most Common Payment method')
a=plt.bar(x.index,x.values,color='y',edgecolor='k')
plt.bar_label(a,padding=2,fontsize=10)
plt.xticks(rotation=45)
```

```python
# 7. What is the most common category of items purchased by female customers with a review rating below 3 ?
import matplotlib.pyplot as plt
import numpy as np
cust2= df.query("Gender =='Female' and `Review Rating`<3 ")
x=cust2['Category'].value_counts()
plt.xlabel('Category')
plt.ylabel('Number of Female customer with rating 3')
plt.title('Most Common Category')
a=plt.bar(x.index,x.values,color='m',edgecolor='k')
plt.bar_label(a,padding=2,fontsize=10)
```

```python
# 8. What is the total purchase amount for customers in each location ?
#df.groupby('Location')['Purchase Amount (USD)'].sum()
import matplotlib.pyplot as plt
import numpy as np
loc1=df.groupby('Location')['Purchase Amount (USD)'].sum()
plt.figure(figsize=(28,9))
x=plt.bar(loc1.index,loc1.values,color='pink',edgecolor='k')
plt.bar_label(x,padding=3,fontsize=10)
plt.xlabel('Total Purchase Amount (USD)')
plt.ylabel('Location')
plt.title('Total Purchase Amount for Customers in Each Location')
plt.xticks(rotation=90)
```

```python
# 9. Plot the correlation between different numerical columns?
import seaborn as sns
import matplotlib.pyplot as plt
column = ['Customer ID', 'Age', 'Purchase Amount (USD)', 'Review Rating', 'Previous Purchases']
numeric_df = df[column]
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
```

```python
# 10. Plot the Histogram of Age Distribution?
import matplotlib.pyplot as plt
import numpy as np
x=df['Age']
plt.xlabel('Age')
plt.ylabel('Number of Customer')
plt.title('Histogram of Age Distribution')
plt.hist(x,width=4,color='r',edgecolor='k')
```

```python
# 11. What is the distribution of customers by age and gender?
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x='Gender', y='Age', data=df)
plt.title('Age Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
```

```python
# 12. Which regions or locations have the highest number of customers?
import matplotlib.pyplot as plt
x= df['Location'].value_counts()
plt.figure(figsize=(24, 9))
plt.barh(x.index,x.values, color='c')
plt.title('Number of Customers by Location', fontsize=16)
plt.xlabel('Number of Customers', fontsize=14)
plt.ylabel('Location', fontsize=14)
```

```python
# 13. What are the most common shopping categories preferred by customers?
import matplotlib.pyplot as plt
x=df['Category'].value_counts()
a=plt.bar(x.index,x.values,color='orange',edgecolor='r')
plt.bar_label(a,padding=2,fontsize=10)
plt.title('Most Common Shopping Categories Preferred by Customers', fontsize=16)
plt.xlabel('Shopping Category', fontsize=14)
plt.ylabel('Number of Customers', fontsize=14)
```

```python
# 14. How does the frequency of shopping vary by age or income group?
import seaborn as sns
import matplotlib.pyplot as plt
frequency_mapping = {'Fortnightly': 1,'Weekly': 2,'Annually': 3,'Quarterly': 4,'Bi-Weekly':5,'Monthly':6,'Every 3 Months':7}
df['Frequency of Purchases'] = df['Frequency of Purchases'].map(frequency_mapping)
sns.lmplot(data=df, x='Age', y='Frequency of Purchases', hue='Gender', palette='Set1', height=6)
plt.title('Age vs. Frequency of Purchases by Gender', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency of Purchases', fontsize=14)
```

```python
# 15. Are there any seasonal trends in shopping behavior?
seasonal_trends = df.groupby(['Season','Gender'])['Purchase Amount (USD)'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(data=seasonal_trends, x='Season', y='Purchase Amount (USD)',hue='Gender',marker='D', markersize=10,palette='Set1')
plt.title('Total Purchase Amount by Season', fontsize=16)
plt.xlabel('Season', fontsize=14)
plt.ylabel('Total Purchase Amount (USD)', fontsize=14)
```
